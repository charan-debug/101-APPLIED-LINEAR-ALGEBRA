```python
# Function to create a matrix
def create_matrix(rows, cols):
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = float(input("Enter value for matrix[{}][{}]: ".format(i, j)))
            row.append(element)
        matrix.append(row)
    return matrix

# Function to add two matrices
def matrix_addition(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix1[0])):
            element = matrix1[i][j] + matrix2[i][j]
            row.append(element)
        result.append(row)
    return result

# Function to subtract two matrices
def matrix_subtraction(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix1[0])):
            element = matrix1[i][j] - matrix2[i][j]
            row.append(element)
        result.append(row)
    return result

# Function to multiply two matrices
def matrix_multiplication(matrix1, matrix2):
    rows1 = len(matrix1)
    cols1 = len(matrix1[0])
    rows2 = len(matrix2)
    cols2 = len(matrix2[0])
    
    # Check if multiplication is possible
    if cols1 != rows2:
        print("Error: Incompatible matrix sizes for multiplication.")
        return
    
    result = []
    for i in range(rows1):
        row = []
        for j in range(cols2):
            element = 0
            for k in range(cols1):
                element += matrix1[i][k] * matrix2[k][j]
            row.append(element)
        result.append(row)
    return result

# Get matrix size from user
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

# Create the first matrix
print("Enter the values for the first matrix:")
matrix1 = create_matrix(rows, cols)

# Create the second matrix
print("Enter the values for the second matrix:")
matrix2 = create_matrix(rows, cols)

# Perform matrix addition
addition_result = matrix_addition(matrix1, matrix2)

# Perform matrix subtraction
subtraction_result = matrix_subtraction(matrix1, matrix2)

# Perform matrix multiplication
multiplication_result = matrix_multiplication(matrix1, matrix2)

# Print the results
print("\nAddition Result:")
for row in addition_result:
    print(row)

print("\nSubtraction Result:")
for row in subtraction_result:
    print(row)

print("\nMultiplication Result:")
for row in multiplication_result:
    print(row)
```
